```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_nba = pd.DataFrame.from_csv('NBA_Season_Data.csv')

def plot_weight(df_nba):
    years = range(1978, 2017)
    for i in range(1,6):
        df_pos = df_nba[df_nba['Rounded Position'] == i]
        df_pos_year = df_pos.groupby(['Year'],axis=0)['Weight'].mean()*0.45359
        plt.plot(years,df_pos_year)
    plt.show()

plot_weight(df_nba)
```
